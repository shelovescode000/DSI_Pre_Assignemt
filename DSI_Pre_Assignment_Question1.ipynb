```python
#load libraries
import math
from random import randint
import numpy as np
import matplotlib.pyplot as plt
```

```python
#function for calculating the traingle of the centroid
def CalculateCentroid(p1, p2, p3):
    pp1 = (p1[0] + p2[0] + p3[0])/3
    pp2 = (p1[1] + p2[1] + p3[1])/3
    return (pp1, pp2)
```

```python
#function for generating the triangle
def GenerateTriangle(p, hex_vertices):
    rand = randint(0, 5)
    a = [-1, 1]
    direction = randint(0,1)
    vert1 = p
    vert2 = hex_vertices[rand]
    pos = rand+a[direction] if rand < 5 else 0
    vert3 = hex_vertices[pos]
    return [vert1, vert2, vert3]
```

```python
#main function: 
def DSI_question1():
    #create a regular hexagon
    hex_vertices = [(-3.52, 2.43), (0.8, 0.99), (4.21, 4.01), (3.29, 8.47), (-1.03, 9.91), (-4.43, 6.89)]
    vals = ['A', 'B', 'C', 'D', 'E', 'F']
    #pick a random point inside the hexagon
    init_p = (1, 6.53)
    x_points = []
    y_points = []
    points = []
    p = init_p
    for i in range(10000):
        tri = GenerateTriangle(p, hex_vertices)
        centroid = CalculateCentroid(*tri)
        points.append(centroid)
        p = centroid
        x_points.append(p[0])
        y_points.append(p[1])
    #plot the scatter plot
    plt.scatter(x_points, y_points)
    plt.show()
```

```python
DSI_question1()
#A beautiful shape a hexagon of hexagons is produced!
```
